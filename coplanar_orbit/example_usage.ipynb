```python
#импортируем класс
from coplanar_orbit import Trajectory
```

```python
#инициализация класса
tj = Trajectory(0,100,1)
tj._spacecraft_param(20_000, 15_000, 20, 3)
tj.trajectory(6500, 0, 0, 7.8, [10,88], [3,4])
```

```python
#визуализируем орбиту
tj.plot(radius_vect=True, dot_engine_start=True)
```

```python
data = tj.data()
data.sample(5)
```

```python
#работа с матрицей состояния 

#максимальная скорости
print(data['|V|'].max())

#апогей и перигей 
#Или в данном случае перигей и наиболее удаленная точка
#Так как КА не дошел до своего апогея

print(data['|r|'].max(), data['|r|'].min())
```
