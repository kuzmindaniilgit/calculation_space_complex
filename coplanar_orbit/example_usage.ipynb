# Построение произвольной орбиты

```python
#импортируем класс
from orbitpy import Trajectory
```

```python
#инициализация класса
tj = Trajectory(0, 135)
tj.spacecraft(20_000, 15_000, 20, 3)
tj.trajectory(6500, 0, 0, 7.8, [52], [5])
```

```python
#визуализируем орбиту
tj.plot(radius_vect=True, dot_engine_start=True)
```

```python
data = tj.data()
data.sample(5)
```

```python
#работа с матрицей состояния 

#максимальная скорости
print(data['|V|'].max())

#апогей и перигей
print(data['|r|'][53:].max(), data['|r|'][53:].min())
```
